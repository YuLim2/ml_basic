```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re

res = urlopen('https://finance.naver.com/sise/') # url 넣기
print(res.status) # 200:성공
soup = BeautifulSoup(res, "html.parser") # 객체 생성

data = soup.find_all('span', class_='num') 
k = re.compile('(\d, {0.1}\d+ \d)')
stocks = k.findall( str(data) )
print( stocks )
for i in range(len(stocks)) :
    stocks[i] = float(  stocks[i].replace(',','')  ) 
print( stocks )
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

serviceK = 'pYzqKPUTp6M7yDoS813%2BXU5ckTuMvKWKZsui9PSsmb7X49iophW18UtvCObQAo7tPt5EvdJZH6k%2FTgNSpWmWHA%3D%3D'

latitude = []
longitude = []
cctvs = []
for pageNum in '1234' :
    url = 'http://apis.data.go.kr/6260000/CctvInfoService/getItsCctvInfoList?serviceKey='+serviceK+'&pageNo='+pageNum+'&numOfRows=10&resultType=xml'
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'lxml')
    #print( soup, type(soup) )

    x = soup.find_all( 'cctv_x' )
    y = soup.find_all( 'cctv_y' )
    n = soup.find_all( 'name' )

    for i in range(len(x)) :
        longitude.append( float(x[i].get_text()) )
        latitude.append( float(y[i].get_text()) )
        cctvs.append( n[i].get_text() )
    
print( latitude )  
print( longitude )
print( cctvs )

df = pd.DataFrame( { 'longitude':longitude, 'latitude':latitude, 'cctvs':cctvs} ) 
print(df.head() )

import folium
from folium.plugins import MarkerCluster
m = folium.Map(
    location = [35.189062673028, 128.90405082258 ],
    zoom_start = 15
)
coords = df[['latitude', 'longitude']]
marker_cluster = MarkerCluster().add_to(m)
for lat, lon in zip( coords['latitude'], coords['longitude']) :
    folium.Marker([lat, lon], icon = folium.Icon(color="green")).add_to(marker_cluster)
m
```
